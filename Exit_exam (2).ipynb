```python
#importing the dataset
d2=pd.read_csv(r"C:\Users\bharath dev\Downloads\Dataset2.csv")
df2=pd.DataFrame(d2)
df2.head()
```

```python
df2.info()
```

```python
df2.isna().sum() #checking for null values
```

```python
df2=df2.dropna() #dropping null rows
```

```python
df2.isna().sum()
```

```python
for col in df2.select_dtypes(include=['number']):  #Plotting boxplot for outlier detection
    plt.figure()
    sns.boxplot(df2[col])
    plt.show()
```

```python
for col in df2.select_dtypes(include=['number']):   #outlier clipping from each outlier columns
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip the outliers
    df2[col] = df2[col].clip(lower=lower_bound, upper=upper_bound)

    




```

```python
for col in df2.select_dtypes(include=['number']):
    plt.figure()
    sns.boxplot(df2[col])
    plt.show()
```

```python
x=df2.drop('Class',axis=1)
```

```python
y=df2['Class']
```

```python
corr=x.corr()   #heatmap of features
plt.figure()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
df2['Class'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder() #Encoding the target variable
le.fit_transform(y)
```

```python
from sklearn.preprocessing import MinMaxScaler  #Scaling numerical cols
scaler=MinMaxScaler()
scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split #Splitting data for training and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier  #Model1: RandomForestClassifier
ensemble=RandomForestClassifier()

ensemble.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix  #evaluating the model

y_pred=ensemble.predict(x_test)
classification_report=classification_report(y_test,y_pred)
conf=confusion_matrix(y_test,y_pred)
print(classification_report)
print(conf)
```

With Random Forest Classifier we got an accuracy score of 0.93 which suggests a good model 

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}


grid_search = GridSearchCV(estimator=ensemble, param_grid=param_grid, 
                           cv=5)


grid_search.fit(x_train, y_train)


best_param = grid_search.best_param_
best_score = grid_search.best_score_
```
